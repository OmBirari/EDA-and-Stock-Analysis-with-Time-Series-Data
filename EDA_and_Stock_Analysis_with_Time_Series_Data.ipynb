```python
import yfinance as yf

# Fetch Tesla's stock data from Yahoo Finance
df_tesla = yf.download('TSLA', start='2016-07-10', end='2023-07-10')

# Display the first few rows of the data
print(df_tesla.head())
```

```python
df_tesla['High'].plot(figsize=(12,4))
```

```python
#x limit and y limit and coloring
df_tesla['High'].plot(xlim=['2021-01-01','2023-01-01'],ylim=[0,700],figsize=(12,4),ls=':',c=('green'))
```

```python
index=df_tesla.loc['2021-01-01':'2023-01-01'].index
share_open=df_tesla.loc['2021-01-01':'2023-01-01']['Open']
```

<h3>Time Resampling

```python
df_tesla.resample(rule='A').min()
```

```python
df_tesla.resample(rule='A').max()['Open'].plot()
```

```python
df_tesla.resample(rule='QS').max()['High'].plot()
```

```python
df_tesla.resample(rule='BQS').max()
```

```python
#plotting
df_tesla['Open'].resample(rule='BA').mean().plot(kind='bar')
```

```python
df_tesla['Open'].resample(rule='M').mean().plot(kind='bar',figsize=(20,8))
```

```python
df_tesla['High'].rolling(10).mean()
```

```python
df_tesla['Open:30 days rolling']= df_tesla['Open'].rolling(30).mean()
```

```python
df_tesla.head(31)
```

```python
df_tesla[['Open','Open:30 days rolling']].plot(figsize=(12,6))
```
